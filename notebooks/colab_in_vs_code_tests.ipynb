```python
print("hello world")

import torch

def get_all_pytorch_devices():
    devices = [torch.device("cpu")]

    # CUDA (NVIDIA GPU / ROCm)
    if torch.cuda.is_available():
        devices += [torch.device(f"cuda:{i}") for i in range(torch.cuda.device_count())]

    # MPS (Apple Silicon)
    if getattr(torch.backends, "mps", None) is not None and torch.backends.mps.is_available():
        devices.append(torch.device("mps"))

    # XLA (TPU, Neuron и др.) — нужен отдельный пакет torch_xla
    try:
        import torch_xla.core.xla_model as xm
        # Например, для TPU: xm.get_xla_supported_devices("TPU")
        xla_devices = xm.get_xla_supported_devices()  # вернёт ['xla:0', 'xla:1', ...]
        devices += [torch.device(d) for d in xla_devices]
    except ImportError:
        # torch_xla не установлен — просто пропускаем
        pass

    return devices
print(get_all_pytorch_devices())

import os
print(os.getcwd())
```
